```python
def see_guts(m):
    
    ####VIZ TRAINING

    fig = plt.figure(figsize=(10, 3))
    ax1 = fig.add_subplot(1,2,1)
    ax2 = fig.add_subplot(1,2,2)
    cmap=plt.get_cmap('Greys')

    fname = os.path.join(m, 'log.json')
    with open(fname, 'r') as f:
        log_all = json.load(f)
    for ri in range(len(rule_trains)):
        r = rule_trains[ri]
        c = cmap((ri+1)/(len(rule_trains)+1))

        x_perf = log_all['perf_'+r]
        x_cost = np.log(log_all['cost_'+r])

        plt.subplot(1,2,1)
        plt.plot(x_perf,'-',c = c)


        plt.subplot(1,2,2)
        plt.plot(x_cost,'-',c = c)

    ax1.set_xlabel('Training Step (x 1000)')
    ax1.set_ylabel('Perf [for each task]')
    ax2.set_xlabel('Training Step (x 1000)')
    ax2.set_ylabel('Log Cost [for each task]')
    plt.show()    

    noise_on = True
    mode = 'test' #'random'
    rule = 'delaygo'

    ####VIZ Inputs, activations and outputs

    model = Model(m)
    with tf.Session() as sess:

        model.restore()#model.saver.restore(sess,ckpt_n_dir)
        # model._sigma=0
        # get all connection weights and biases as tensorflow variables
        var_list = model.var_list
        # evaluate the parameters after training
        params = [sess.run(var) for var in var_list]
        # get hparams
        hp = model.hp

        trial = generate_trials(rule, hp, mode, batch_size=100, noise_on = noise_on)
        trial_no_noise = generate_trials(rule, hp, mode, batch_size=100, noise_on = False)

        # Generating feed_dict.
        feed_dict = tools.gen_feed_dict(model, trial, hp)
        h_tf, y_hat_tf = sess.run([model.h, model.y_hat], feed_dict=feed_dict) 

    plt.plot(trial.x[:,0,:],alpha = .5)
    plt.plot(trial_no_noise.x[:,0,:],'--',alpha = 1,linewidth = 4)

    plt.xlabel('timesteps')
    plt.ylabel('activation (inputs)')
    plt.show()

    plt.plot(h_tf[:,1,:]) #T time x B trials x N neurons

    plt.xlabel('timesteps')
    plt.ylabel('activation (hidden units)')
    plt.show()

    c_set = ['orangered','dodgerblue','k']
    for out_i in range(3):
        plt.plot(trial.y[:,1,out_i], c = c_set[out_i],linewidth = 4,alpha = .5) #T time x B trials x N neurons
        plt.plot(y_hat_tf[:,1,out_i], c = c_set[out_i],linewidth = 2,alpha = .5) #T time x B trials x N neurons

    plt.xlabel('timesteps')
    plt.ylabel('activation (outputs)')
    plt.show()
    return hp
```
